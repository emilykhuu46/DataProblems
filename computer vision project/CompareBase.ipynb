# Compare models that were trained on the original unaugmented image data

**Load image data**

```python

import os
import numpy as np
import keras
from keras import layers
from tensorflow import data as tf_data
import matplotlib.pyplot as plt

import DataLoader

train_ds, val_ds = DataLoader.load_image_dataset("reptile_amphibian_pics")

DataLoader.show_images(train_ds)
```

# Load MobileNet Model

```python
from keras.models import load_model

# Load saved model
model_1 = load_model("my_model.keras")
```

```python
history = model_1.fit(train_ds, epochs=8, validation_data=val_ds)
```

```python

plt.plot(range(len(history.history["loss"])),history.history["loss"],label="Training Loss")
plt.plot(range(len(history.history["val_loss"])),history.history["val_loss"],label="Validation Loss")
plt.legend()
```

```python
plt.plot(range(len(history.history["acc"])),history.history["acc"],label="Training Accuracy")
plt.plot(range(len(history.history["val_acc"])),history.history["val_acc"],label="Validation Accuracy")
plt.legend()
```

# Load MobileNetV2

```python
model_2 = load_model("my_model3.keras")
```

```python
history2 = model_2.fit(train_ds, epochs=8, validation_data=val_ds)
```

```python
plt.plot(range(len(history2.history["loss"])),history2.history["loss"],label="Training Loss")
plt.plot(range(len(history2.history["val_loss"])),history2.history["val_loss"],label="Validation Loss")
plt.legend()
```

# Comparisons

**MobileNet**  
final training accuracy: 0.9377 and training loss: 0.1749  
final val accuracy: 0.9200 and val loss: 0.2283

+ This model has good accuracy and low loss values. The loss and accuracy curves are also relatively parallel to each other with a couple of small jumps here and there. It is performing pretty well.

**MobileNetV2**  
final training accuracy: 0.9552 and training loss: 0.1305  
final val accuracy: 0.9150 and val loss: 0.2242

+ For this model the accuracy and loss are even better than the first mobilenet. The curves are also more parallel and has less jumps than mobilenet. 

**InceptionV3**  
final training accuracy: 0.9564 and training loss: 0.1107  
final val accuracy: 0.9350 and val loss: 0.1686

+ For this model the accuracy and loss are even better than the mobilenet2. The curves show that these are parallel and smoother than the past two models and has almost no instances of sharp jumps.


Overall, MobileNet performed the worst, then MobileNetV2 did better, and InceptionV3 did the best. More tuning could be done on the tuning on all of the models to get better results, but the results were high for all of them anyways. 

InceptionV3 >> MobileNetV2 >> MobileNet

**The InceptionV3 transfer learning model performed the best as it had the lowest loss and the highest accuract for both training and validation as well as had the more parallel and flattest curve.**
